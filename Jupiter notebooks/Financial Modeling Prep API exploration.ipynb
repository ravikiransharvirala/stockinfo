```python
data = data[data['exchange'].isin(['NYSE', 'Nasdaq', 'New York Stock Exchange'])]
```

```python
profile_data = []
profile_keys = ['symbol', 'price', 'mktCap', 'sector', 'industry']
for s in data['symbol']:
    url = "https://financialmodelingprep.com/api/v3/profile/{}?apikey=e411bd9e9db8094b51a796cc4a6345fc".format(s)
    r = requests.get(url)
    req_data = r.json()[0]
    req_data = {key : req_data[key] for key in profile_keys}
    profile_data.append(req_data)
```

```python
p_data = pd.DataFrame(profile_data)
```

```python
data.merge(p_data, on='symbol', how='inner')
```
